```python
import matplotlib.pyplot as plt
import numpy as np

# Generate sample data
np.random.seed(42)
x = np.random.rand(50)
y = np.random.rand(50)
colors = np.random.rand(50)

# Create a scatter plot
scatter = plt.scatter(x, y, c=colors, cmap='viridis', label='Variable')

# Add colorbar
cbar = plt.colorbar(scatter, label='Variable Scale')

# Add legend
legend_labels = ["Low", "Medium", "High"]
cbar.set_ticks([0, 0.5, 1])
cbar.set_ticklabels(legend_labels)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot with Color Variable')

# Show the plot
plt.show()
```
